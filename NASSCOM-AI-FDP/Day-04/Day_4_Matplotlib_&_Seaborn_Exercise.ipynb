```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("--- Intermediate Matplotlib & Seaborn Exercise ---")
print("Complete each task by writing the requested plotting code.")
print("Focus on proper labeling, titles, and drawing insights.")
print("--------------------------------------------------")
```

```python
tips = sns.load_dataset('tips')
display(tips.head())
```

```python
# 1.1 Histogram: Distribution of 'total_bill'
# Create a histogram of the 'total_bill' column.
# Add a title, x-label, and y-label. Use `bins=20`.
# Insight Question: What is the typical range of total bills? Is the distribution symmetric or skewed?

plt.figure(figsize=(10, 6))
sns.histplot(tips['total_bill'], bins=20, kde=True)
plt.title('Distribution of Total Bills')
plt.xlabel('Total Bill ($)')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

### Insight:
From the histogram, the typical range of total bills appears to be between $10 and $20, with a peak frequency around $15-$18. The distribution is clearly skewed to the right (positively skewed), meaning there are a few higher total bills, but most bills are on the lower end.

```python
smoker_counts = tips['smoker'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(smoker_counts, labels=smoker_counts.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral'])
plt.title('Proportion of Smokers vs. Non-Smokers')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

### Insight:
The pie chart clearly shows that the majority of customers in this dataset are non-smokers, accounting for 61.3% of the total. Smokers make up a significant, but smaller, portion at 38.7%. This indicates a notable difference, with non-smokers being almost twice as prevalent as smokers in this dataset.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=tips, x='day', y='tip', palette='viridis')
plt.title('Tip Amount Distribution by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Tip Amount ($)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### Insight:
From the box plot, we can observe differences in tip amounts across the days. Sunday and Saturday appear to have higher median tip amounts compared to Thursday and Friday. Saturday also shows a wider interquartile range (IQR) and more outliers, indicating a greater variability in tip amounts on that day. Thursday seems to have the lowest median tip amount and a narrower distribution. The outliers suggest that on any given day, there can be exceptionally high tips, but they are more frequent on weekends.

```python
plt.figure(figsize=(12, 7))
sns.barplot(data=tips, x='day', y='total_bill', hue='time', palette='coolwarm', errorbar=None) # errorbar=None removes error bars for cleaner visualization
plt.title('Average Total Bill by Day and Time')
plt.xlabel('Day of the Week')
plt.ylabel('Average Total Bill ($)')
plt.legend(title='Time')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### Insight:
The grouped bar chart reveals clear patterns in average total bills. Dinner bills are consistently higher than lunch bills across all days. On Thursday and Friday, lunch is served, and dinner bills are notably higher. On Saturday and Sunday, only dinner is served, and these days generally show the highest average total bills overall. This suggests that customers tend to spend more during dinner service, and weekend dinners are particularly more expensive on average.

```python
numerical_data = tips[['total_bill', 'tip', 'size']]
correlation_matrix = numerical_data.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

### Insight:
The heatmap clearly shows the correlation between the numerical features:
- **Total Bill and Tip:** There is a strong positive correlation (0.68) between `total_bill` and `tip`, indicating that as the total bill increases, the tip amount generally increases as well. This confirms our earlier observation from the scatter plot.
- **Total Bill and Size:** There is a moderate positive correlation (0.60) between `total_bill` and `size`, suggesting that larger parties tend to have larger total bills.
- **Tip and Size:** There is also a moderate positive correlation (0.49) between `tip` and `size`, implying that larger parties tend to leave higher tips, which is expected given their larger total bills.

```python
# Create a figure with 2 rows and 2 columns of subplots.
#    - Top-left: Histogram of 'tip' amount.
#    - Top-right: Box plot of 'total_bill' by 'sex'.
#    - Bottom-left: Count plot of 'day'.
#    - Bottom-right: Swarm plot of 'tip' by 'smoker' status.
# Ensure each subplot has an appropriate title and labels. Use `plt.tight_layout()` to prevent overlap.
# Insight Question: What can you learn about tip distribution, total bill distribution across genders, daily customer patterns, and tip differences based on smoker status from these combined plots?

fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Top-left: Histogram of 'tip' amount
sns.histplot(tips['tip'], bins=15, kde=True, ax=axes[0, 0])
axes[0, 0].set_title('Distribution of Tip Amount')
axes[0, 0].set_xlabel('Tip Amount ($)')
axes[0, 0].set_ylabel('Frequency')

# Top-right: Box plot of 'total_bill' by 'sex'
sns.boxplot(x='sex', y='total_bill', data=tips, ax=axes[0, 1], palette='pastel')
axes[0, 1].set_title('Total Bill Distribution by Sex')
axes[0, 1].set_xlabel('Sex')
axes[0, 1].set_ylabel('Total Bill ($)')

# Bottom-left: Count plot of 'day'
sns.countplot(x='day', data=tips, ax=axes[1, 0], palette='viridis')
axes[1, 0].set_title('Customer Count by Day of the Week')
axes[1, 0].set_xlabel('Day of the Week')
axes[1, 0].set_ylabel('Number of Customers')

# Bottom-right: Swarm plot of 'tip' by 'smoker' status
sns.swarmplot(x='smoker', y='tip', data=tips, ax=axes[1, 1], palette='muted')
axes[1, 1].set_title('Tip Amount by Smoker Status')
axes[1, 1].set_xlabel('Smoker')
axes[1, 1].set_ylabel('Tip Amount ($)')

plt.tight_layout()
plt.show()
```

### Combined Insights:
This combined set of plots offers several insights:
- **Tip Distribution:** The histogram shows that tip amounts are primarily concentrated between $2 and $4, with a right-skewed distribution, indicating a tendency for smaller tips with occasional higher tips.
- **Total Bill by Sex:** The box plot suggests that while the median total bill for males and females is similar, males tend to have a slightly wider range and higher outliers for total bills, implying they might occasionally incur larger bills than females.
- **Daily Customer Patterns:** The count plot reveals that Saturday and Sunday are the busiest days in terms of customer count, followed by Thursday. Friday has the lowest customer count among the days recorded.
- **Tip by Smoker Status:** The swarm plot indicates that both smokers and non-smokers show a similar range of tip amounts. While non-smokers appear to have a denser cluster of tips in the lower to mid-range, there isn't a strikingly clear difference in the overall tip amount distribution or maximum tips between the two groups. Both groups show outliers with higher tip amounts.
